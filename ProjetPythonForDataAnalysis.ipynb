# Import & Preprocessing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
datatr = pd.read_csv('avila-tr.txt', sep=",", header=None)
datatr.columns = ["intercolumnar_distance","upper_margin","lower_margin","exploitation","row_number","modular_ratio","interlinear_spacing","weight","peak_number","modular_ratio/interlinear_spacing","Class"]
datatr = datatr[(datatr.modular_ratio!=53)]
datatr
```

```python
datats = pd.read_csv('avila-ts.txt', sep=",", header=None)
datats.columns = ["intercolumnar_distance","upper_margin","lower_margin","exploitation","row_number","modular_ratio","interlinear_spacing","weight","peak_number","modular_ratio/interlinear_spacing","Class"]
datats
```

```python
datatr.describe().loc[['mean', 'std', 'min', 'max']]
```

Here, we can see that an unique value is different from the others, and as it is influencing means too much, we decided to exclude it from our analysis.

```python
datatr.sort_values(by="lower_margin",ascending=False)
```

According to the associated research paper :
"A Method for Scribe Distinction in Medieval Manuscripts Using Page Layout Features"
The parameters 'modular_ratio' and 'modular_ratio/interlinear_spacing' prove to be cumbersome and misleading, and do not help to the classification of our dataset. As such, we decided to exclude them.

```python
data = pd.concat([datatr,datats])
data = data.drop(labels='modular_ratio/interlinear_spacing', axis=1)
data = data.drop(labels='modular_ratio', axis=1)
data
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=data.drop(labels='Class', axis=1)
y=data.Class
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.2, random_state=1) #test_size=4000

Xtrain.shape, Xtest.shape
```

```python
train, test = train_test_split(data, test_size=0.2)

train.shape, test.shape
```

# Machine Learning

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF

from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
```

## RandomForestClassifier

```python
param_grid_RForest = {'n_estimators': np.arange(10,201,40),
                      'criterion':['gini', 'entropy'],
                      #'min_samples_split': np.arange(1,6,1),
                      #'min_samples_leaf': np.arange(1,6,1),
                      'min_weight_fraction_leaf': np.arange(0,0.6,0.1),
                      'max_features':['sqrt','log2'], 
                      #'bootstrap':[True, False],
                      #'oob_score':[True, False],
                      'n_jobs': np.arange(-1,2,2),
                      'verbose': [1]}

grid_RForest = GridSearchCV(RandomForestClassifier(), param_grid_RForest, cv=5)

grid_RForest.fit(X, y);

grid_RForest.best_params_
```

```python
model_RForest = grid_RForest.best_estimator_
score_RForest = cross_val_score(model_RForest, X, y)
avg_score_RForest = score_RForest.mean()
print(avg_score_RForest)
```

```python
Xtr=datatr # X is standard name for features matrix
ytr=datatr["Class"] # y is standard name for target array
Xtr = Xtr.drop(labels='Class', axis=1)
Xts=datats # X is standard name for features matrix
yts=datats["Class"] # y is standard name for target array
Xts = Xts.drop(labels='Class', axis=1)
```

```python
grid_RForest_normal = GridSearchCV(RandomForestClassifier(), param_grid_RForest, cv=5)

grid_RForest_normal.fit(Xtr, ytr);

grid_RForest_normal.best_params_
```

```python
model_RForest_normal = grid_RForest_normal.best_estimator_
score_RForest_normal = cross_val_score(model_RForest_normal, Xts, yts)
avg_score_RForest_normal = score_RForest_normal.mean()
print(avg_score_RForest_normal)
```

```python
y_model_RForest_normal = model_RForest_normal.predict(Xts)
from sklearn.metrics import accuracy_score
accuracy_score(yts, y_model_RForest_normal)
```

Here, we compare the accuracy resulting from both methods, with cross validation (avg_score_RForest), and with the split dataset : training & testing (avg_score_RForest_normal).

```python
RForest_score_difference = round(((avg_score_RForest - avg_score_RForest_normal) * 100) / avg_score_RForest_normal, 2)

print(f'Ancienne précision:\t', avg_score_RForest_normal)
print(f'Nouvelle précision:\t', avg_score_RForest, f'({RForest_score_difference}%)')
```

Our goal here was to test all parameters, one at a time, since it took more than 3 days to complete a GridSearch analysis.

The algorithm used (ball tree) did not change since it is the best suited for our needs.

The leaf_size seems to have no impact on the accuracy. 
We tried with values as extremes as 1 and 10000 and the accuracy stayed substantially constant.

We realised that a higher P value had negative impact on the accuracy.

## SVC

```python
#Basic version of the SVC model
#It is recommended to using the standard normal distribution, but our data have already gone throught such process.
base_model_svc = SVC()
base_score_svc = cross_val_score(base_model_svc, X, y, cv=5)
base_score_avg_svc = base_score_svc.mean()
base_score_avg_svc
```

Our goal here was to test all parameters, one at a time, since it took more than a day to complete a GridSearch analysis. 
It isn't as accurate as running it with all the parameters simultaneously, but it is way faster.

First we tested the different types of kernel.

As the best algorithm was found to be rbf, we have been able to remove parameters seuch as degree and coef0 which were only used for the poly and sigmoid algorithms.

Then we tested different values for the paramter C, obtaining ever better result as C grew, but as this change would overfit our model, we decided to keep it a 100 (being 1.0 by default).

The parameters shrinking and probability didn't increase significantly our accuracy and made the computation of the model longer by a factor of 10 (taking half an hour).

Finally, we saw that a greater value for the tolerance parameter increased the accuracy.

```python
param_grid_svc = {'C': [1],
              'kernel':['linear', 'poly', 'rbf', 'sigmoid'],
              'degree':[3],
              'gamma':['scale'],
              'coef0':[0],
              'shrinking':[True],
              'probability':[False],
              'tol':[1e-3]}

grid_svc = GridSearchCV(SVC(), param_grid_svc, cv=5)
grid_svc.fit(X, y);
print(grid_svc.best_params_)
model_svc = grid_svc.best_estimator_
score_svc = cross_val_score(model_svc, X, y)
avg_score_svc = score_svc.mean()
print(avg_score_svc)
```

```python
from sklearn.model_selection import validation_curve
```

```python
param_range_n_estimators = np.arange(10,201,40)
train_scores_RForest, test_scores_RForest = validation_curve(
    RandomForestClassifier(), X, y, param_name="n_estimators", param_range=param_range_n_estimators,
    scoring="accuracy", n_jobs=1)
train_scores_mean_RForest = np.mean(train_scores_RForest, axis=1)
train_scores_std_RForest = np.std(train_scores_RForest, axis=1)
test_scores_mean_RForest = np.mean(test_scores_RForest, axis=1)
test_scores_std_RForest = np.std(test_scores_RForest, axis=1)

plt.figure(figsize=(9, 9))
plt.title("Validation Curve with _RForest")
plt.xlabel("n_estimators")
plt.ylabel("Score")
plt.ylim(0.0, 1.1)
lw = 2
plt.semilogx(param_range_n_estimators, train_scores_mean_RForest, label="Training score",
             color="darkorange", lw=lw)
plt.fill_between(param_range_n_estimators, train_scores_mean_RForest - train_scores_std_RForest,
                 train_scores_mean_RForest + train_scores_std_RForest, alpha=0.2,
                 color="darkorange", lw=lw)
plt.semilogx(param_range_n_estimators, test_scores_mean_RForest, label="Cross-validation score",
             color="navy", lw=lw)
plt.fill_between(param_range_n_estimators, test_scores_mean_RForest - test_scores_std_RForest,
                 test_scores_mean_RForest + test_scores_std_RForest, alpha=0.2,
                 color="navy", lw=lw)
plt.legend(loc="best")
plt.show()
```

```python
param_range_min_weight_fraction_leaf = np.arange(0,0.6,0.1)
train_scores_RForest, test_scores_RForest = validation_curve(
    RandomForestClassifier(), X, y, param_name="min_weight_fraction_leaf", param_range=param_range_min_weight_fraction_leaf,
    scoring="accuracy", n_jobs=1)
train_scores_mean_RForest = np.mean(train_scores_RForest, axis=1)
train_scores_std_RForest = np.std(train_scores_RForest, axis=1)
test_scores_mean_RForest = np.mean(test_scores_RForest, axis=1)
test_scores_std_RForest = np.std(test_scores_RForest, axis=1)

plt.figure(figsize=(9, 9))
plt.title("Validation Curve with _RForest")
plt.xlabel("min_weight_fraction_leaf")
plt.ylabel("Score")
plt.ylim(0.0, 1.1)
lw = 2
plt.semilogx(param_range_min_weight_fraction_leaf, train_scores_mean_RForest, label="Training score",
             color="darkorange", lw=lw)
plt.fill_between(param_range_min_weight_fraction_leaf, train_scores_mean_RForest - train_scores_std_RForest,
                 train_scores_mean_RForest + train_scores_std_RForest, alpha=0.2,
                 color="darkorange", lw=lw)
plt.semilogx(param_range_min_weight_fraction_leaf, test_scores_mean_RForest, label="Cross-validation score",
             color="navy", lw=lw)
plt.fill_between(param_range_min_weight_fraction_leaf, test_scores_mean_RForest - test_scores_std_RForest,
                 test_scores_mean_RForest + test_scores_std_RForest, alpha=0.2,
                 color="navy", lw=lw)
plt.legend(loc="best")
plt.show()
```
